## Additional Information
- <mark>Technical indicators</mark> (often called "tech index" in short) are mathematical calculations based on an asset’s price, volume, or open interest. They are widely used in technical analysis to forecast future price movements, identify trends, and generate buy/sell signals.

These indicators help traders make informed decisions by:
- Identifying trends (upward, downward, or sideways)
- Measuring momentum and strength of price moves
- Detecting overbought or oversold conditions
- Spotting potential reversals or breakouts
- Confirming trading signals from price patterns
Technical indicators are typically applied to charts (e.g., candlestick charts) and can be leading (predict future price moves) or lagging (follow price action).

### Rolling windows

<mark>Pandas rolling() function</mark> gives the element of moving window counts. 

- The idea of moving window figuring is most essentially utilized in signal handling and time arrangement information.  The window of size k implies k back-to-back qualities one after another. All the ‘k’ values are similarly weighted in an entire case.

```python
import yfinance as yf
# Download BTC-USD data
data = yf.Ticker("BTC-USD")
df = data.history(start="2024-01-01",end="2030-12-31").tz_localize(None) # importance
df = df.drop(['Dividends', 'Stock Splits'], axis=1)
df.dropna(inplace=True)

df.info()
df.tail(20)
```

```python
import pandas as pd
# 括號內代表要滾動幾個資料點的參數。要計算平均數，我們後面加mean()。
df['SMA5']=df['Close'].rolling(5).mean() # <-- 5日移動平均
df['SMA20']=df['Close'].rolling(20).max() # <--- 計算滾動20日高點
df['SMA5'].dropna(inplace=True)
df['SMA20'].dropna(inplace=True)
```

#### 1. Exponential Moving Average
- Purpose: Smooths price data to identify trends.
- Types: Simple Moving Average (SMA) and Exponential Moving Average (EMA)

We can calculate exponential moving averages using ewm functions. EMA is a technical indicator which help us to determine the direction of a stock movement based on the past prices. if we say 9 ema, then the moving average of past 9 candles are considered.

- Crossovers (e.g., EMA 50 vs EMA 200 – “Golden Cross” or “Death Cross”)

```python
df['EMA5'] = df['Close'].ewm(span=5, adjust=False).mean()
df['EMA20'] = df['Close'].ewm(span=20, adjust=False).mean()
df['EMA5'].dropna(inplace=True)
df['EMA20'].dropna(inplace=True)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(14, 7))
plt.plot(df['Close'], label='Closing Price', color='blue')
#plt.plot(df['SMA5'], label='5 Day MA', color='red')
#plt.plot(df['SMA20'], label='20 Day MA', color='grey')

plt.plot(df['EMA5'], label='5 Day EMA', color='pink')
plt.plot(df['EMA20'], label='20 Day EMA', color='black')

plt.title('MA vs EMA', fontsize=20)
plt.ylabel('Price', fontsize=20)
plt.xlabel('Dates', fontsize=15)
plt.legend()
plt.show()
```

#### 2. MACD Moving Average Convergence/Divergence
- Purpose: Measures momentum and trend direction.
- Components: MACD line, Signal line, Histogram

MACD is derived from two Exponential Moving Averages (EMAs): a short-term (12 priods) EMA and a long-term (26 periods) EMA. Then we can calculus the MACD = short-period closing price EMA – long-period closing price EMA. After subtracting the EMA’s, we need to calculate the signal line. The signal line is the 9-period EMA .

In the past, the working week was 6 days, so the period settings of (12, 26, 9) represented 2 weeks, 1 month and 1½ weeks respectively. Now, when the trading week is only 5 days, the period parameters can be adjusted accordingly, e.g. (10, 20, 7).

Signals:
- MACD line crosses above signal line → Buy
- MACD line crosses below → Sell
- Divergence with price → potential reversal

```python
# Calculate the 12-period and 26-period EMA
df['EMA12'] = df['Close'].ewm(span=12, adjust=False).mean()
df['EMA26'] = df['Close'].ewm(span=26, adjust=False).mean()

# Calculate MACD (the difference between 12-period EMA and 26-period EMA)
df['MACD'] = df['EMA12'] - df['EMA26']

# Calculate the 9-period EMA of MACD (Signal Line)
df['Signal_Line'] = df['MACD'].ewm(span=9, adjust=False).mean()
```

```python
# Plotting Moving Average Convergence Divergence
plt.figure(figsize=(14, 7))
plt.plot(df['MACD'], label='MACD Line', color='red')
plt.plot(df['Signal_Line'], label='Signal Line', color='blue')
plt.bar(df.index, df['MACD'], label='MACD Histogram', color='grey')
plt.title('Moving Average Convergence Divergence')
plt.legend()
plt.show()
```

#### 3. RSI Relative Strength Index
- Purpose: Measures speed and change of price movements (momentum oscillator).
- Range: 0 to 100

The Relative Strength Index (RSI) was first introduced by Welles Wilder in 1978. It is a momentum oscillator that measures the strength of recent asset price movements. It is usually calculated using 14 days of price data. The principle of the RSI is to compare the average number of days an asset has risen in the recent past with the average number of days it has fallen.

$$
RSI = 100 - \frac{100}{1+\text{RS}}, \text{RS}=\frac{\text{Average Gain}}{\text{Average Loss}}
$$

Signals:
- RSI > 70 → Overbought (possible sell)
- RSI < 30 → Oversold (possible buy)
- RSI divergence → potential trend reversal

```python
change = df['Close'].diff()
change.dropna(inplace=True)

# Create two copies of the Closing price Series
change_up = change.copy()
change_down = change.copy()

change_up[change_up < 0] = 0
change_down[change_down > 0] = 0

# Calculate the rolling average of average up and average down
avg_up = change_up.rolling(14).mean()
avg_down = change_down.rolling(14).mean().abs()
# rs = avg_up / avg_down

df['RSI_14'] = 100 * avg_up / (avg_up + avg_down)
# rsi = 100 -(100/ (1 + rs))
```

```python
fig, ax = plt.subplots(figsize=(14, 7))
df["RSI_14"].plot(ax=ax, title="Relative Strength Index (RSI)")
ax.hlines(y=30, xmin=df.index.min(),
          xmax=df.index.max(), color="red")
ax.hlines(y=70, xmin=df.index.min(),
          xmax=df.index.max(), color="red")
plt.show()
```

#### 4. KDJ Indicator
- Purpose: Compare a closing price to its price range over time.
- Range: 0–100

KDJ indicator is known as the random index. KDJ is a derived form of the <mark>Stochastic Oscillator Indicator</mark> with the only difference of having an extra line called the J line. The J line represents the divergence of the %D value from the %K. The value of J can go beyond [0, 100] for %K and %D lines on the chart.

- RSVt=(Ct-L9)/(H9-L9)*100 (Ct=closing price of the day; L9=lowest price within 9 days; H9=highest price within 9 days)
- The K value is the 3-day smoothed moving average of the RSV value. 
- The D value is the 3-day smoothed moving average of the K value. 
- The J value is three times the K value minus two times the D value.

Signals of Stochastic Oscillator Indicator
- 80 → overbought 
- <20 → oversold
- %K crosses %D → momentum shift


```python
low_list = df['Low'].rolling(9).min()
low_list.fillna(value=df['Low'].expanding().min(), inplace=True)
high_list = df['High'].rolling(9).max()
high_list.fillna(value=df['High'].expanding().max(), inplace=True)

rsv = (df['Close']-low_list) / (high_list - low_list)* 100
df['KDJ_k'] = rsv.ewm(com=2).mean()
df['KDJ_D'] = df['KDJ_k'].ewm(com=2).mean()
df['KDJ_J'] = 3 * df['KDJ_k'] - 2 * df['KDJ_D']
```

```python
df[["Close", "KDJ_k", "KDJ_D", "KDJ_J"]].plot(title="KDJ" , figsize=(12, 6))
```

#### 5. Bollinger Bands
- Components: Middle (SMA), Upper & Lower bands (±2 standard deviations)
- Purpose: Measure volatility and relative price levels.

John Bollinger in the 1980s developed them and are very simple and intuitive to use. These bands offera unique perspective on volatility. These bnands consists three lines:

Middle Band (MA), Upper Band (UB): This is the sum of the 20-period SMA and two times the 20-period standard deviation (SD) of the closing prices. Lower Band (LB): This is the 20-period SMA minus two times the 20-period SD of the closing prices.

Signals:
- Price touches upper band → overbought
- Price touches lower band → oversold
- “Bollinger Squeeze” → low volatility, potential breakout

```python
# Calculate the 20-period Simple Moving Average (SMA)
df['SMA20'] = df['Close'].rolling(window=20).mean()
# Calculate the 20-period Standard Deviation (SD)
df['SD20'] = df['Close'].rolling(window=20).std()

# Calculate the Upper Bollinger Band (UB) and Lower Bollinger Band (LB)
df['UB'] = df['SMA20'] + 2 * df['SD20']
df['LB'] = df['SMA20'] - 2 * df['SD20']
```

```python
df[["Close", "UB", "LB"]].plot(title="Bollinger Band", figsize=(12, 6))
```

## Another useful tech index
Ichimoku Cloud (Ichimoku Kinko Hyo)
-Purpose: Comprehensive indicator showing trend, momentum, and support/resistance.
Components:
Tenkan-sen, Kijun-sen, Senkou Span A/B (forms the “cloud”)

Use Case:
- Price above cloud → bullish
- Price below cloud → bearish
- Cloud color change → trend shift

Average True Range (ATR)
- Purpose: Measures market volatility.
Not directional – only shows how much an asset moves.

Use Case:
- Set stop-loss levels
- Filter low-volatility assets

Volume Weighted Average Price (VWAP)
- Purpose: Average price weighted by volume (popular in intraday trading).

Use Case:
- Price above VWAP → bullish bias
- Price below VWAP → bearish bias
- Institutional traders use it as a benchmark

Fibonacci Retracement
- Purpose: Identify potential support/resistance levels based on Fibonacci ratios (23.6%, 38.2%, 50%, 61.8%).

Use Case:
- Pullbacks in a trend
- Entry/exit points during corrections

On-Balance Volume (OBV)
- Purpose: Track cumulative volume flow to confirm price trends.

Logic:
- Volume added on up days, subtracted on down days
- Rising OBV → bullish momentum
- Divergence → potential reversal
